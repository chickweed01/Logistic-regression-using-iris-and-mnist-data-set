```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np
```

```python
mnist = input_data.read_data_sets("MNIST_DATASET/", one_hot=True)
```

```python
with tf.name_scope("Declaring_hyperparametre"):
    learning_rate = 0.01
    num_epochs = 1000
    batch_size = 100
    display_step = 1
with tf.name_scope("Declaring_placeholders"):
    X = tf.placeholder(tf.float32, [None, 784])
    y = tf.placeholder(tf.float32, [None, 10])
with tf.name_scope("Declaring_variables"):
    W = tf.Variable(tf.zeros([784, 10]))
    b = tf.Variable(tf.zeros([10]))
with tf.name_scope("Declaring_model_function"):
    prediction = tf.nn.softmax(tf.add(tf.matmul(X, W), b))
```

```python
with tf.name_scope("Declaring_cost_function"):
    cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y, logits=prediction))
with tf.name_scope("Declaring_optimizer"):
    optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
with tf.name_scope("starting_session"):
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        for epoch in range(num_epochs):
            total_batch = int(mnist.train.num_examples/batch_size)
            cost_in_epoch = 0
            for i in range(total_batch):
                batch_X, batch_y = mnist.train.next_batch(100)
                _, c = sess.run([optimizer, cost], feed_dict={X: batch_X, y: batch_y})
                cost_in_epoch += c / total_batch
            if (epoch+1) % display_step == 0:
                print("Epoch:", '%04d' % (epoch+1), "cost=", "{:.9f}".format(cost_in_epoch))

            
        print("Optimization Finished!")

        correct_prediction = tf.equal(tf.argmax(prediction, 1), tf.argmax(y, 1))
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
        print("Accuracy:", accuracy.eval({X: mnist.test.images[:3000], y: mnist.test.labels[:3000]}))
```
